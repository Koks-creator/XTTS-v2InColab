```python
%%writefile /content/run_tts.py
import os
os.environ["MPLBACKEND"] = "Agg"
os.environ["COQUI_TOS_AGREED"] = "1"

import torch
import torchaudio
import soundfile as sf
import numpy as np
from datetime import datetime

MOUNT_PATH = "/content/drive/MyDrive/xtts_szpont/"
MOUNT_DIR = os.path.dirname(MOUNT_PATH)
VOICES_DIR = f"{MOUNT_DIR}/voices"
OUTPUT_DIR = f"{MOUNT_DIR}/output"


_orig_torch_load = torch.load
def _patched_torch_load(*args, **kwargs):
    kwargs.setdefault("weights_only", False)
    return _orig_torch_load(*args, **kwargs)
torch.load = _patched_torch_load

def _patched_audio_load(path, *args, **kwargs):
    data, sr = sf.read(str(path), dtype="float32", always_2d=True)
    # soundfile: (samples, channels) в†’ torchaudio: (channels, samples)
    tensor = torch.from_numpy(data.T).contiguous()
    return tensor, sr
torchaudio.load = _patched_audio_load

from TTS.api import TTS

device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

# Wir fГјhren bei Colab Turbotests des Modells XTTS durch
# Wielkie testy modelu XTTS w colabku tak o
# let's test XTTS in colab, just like that
tts.tts_to_file(
    text="As i once told you, my late uncle, i was very young. I was maybe around 5 or 6 years old, i don't even know, i was very young my uncle was at my grandmother's and AS A JOKE I EMPHASIZE AS A JOKE I rhymed at my uncle 'Uncle-Knucle' with a K, with a K because it rhymed, with a K. I said 'uncle-kuncle' and then uncle slapped the shit out of me and since that time i have never made a single joke towards my uncle again.",
    file_path=f"{OUTPUT_DIR}/output_{datetime.strftime(datetime.now(), '%Y_%m_%d_%H_%M_%S')}.mp3",
    speaker_wav=f"{VOICES_DIR}/audio_2026-05-09_22-58-58.ogg",
    language="en",
)
print("DONE")
```
